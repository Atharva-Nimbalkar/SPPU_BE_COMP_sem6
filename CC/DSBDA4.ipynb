```python
X=df[['crim','zn','indus','chas','nox','rm','age','dis','rad','tax','ptratio','b','lstat']]
Y=df[['medv']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
```

Use StandardScaler if you want each feature to have zero-mean, unit standard-deviation. If you want more normally distributed data, and are okay with transforming your data.

Normalization is the process of scaling individual samples to have unit norm. This process can be useful if you plan to use a quadratic form such as the dot-product or any other kernel to quantify the similarity of any pair of samples.

Regularization is a set of methods for reducing overfitting in machine learning models.

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)

# print(y_pred)
```

Output:

```python
# Create a DataFrame for the output
y_pred = lr.predict(X_test)
print(y_pred)
```

```python
#evaluate


from sklearn.metrics import mean_squared_error,r2_score
mse = mean_squared_error(y_test, y_pred)
r2=r2_score(y_test,y_pred)
print("Mean squared Error is:")
print(mse)
print("r2 squared Error is:")
print(r2)
```

```python
print("Training accuracy is:")
# The best possible score is 1.0 and it can be negative (because the
# model can be arbitrarily worse). A constant model that always predicts
# the expected value of y, disregarding the input features, would get
# a R^2 score of 0.
lr.score(X_train, y_train)
```

```python
print("Testing accuracy is:")
lr.score(X_test, y_test)
```

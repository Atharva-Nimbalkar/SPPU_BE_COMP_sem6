```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
df=sns.load_dataset('titanic')
df
```

```python
plt.figure(figsize=(16,8))
sns.countplot(data=df,x='age')
plt.xticks(rotation=90)
plt.xlabel("Age")
plt.ylabel("Count")
```

```python
plt.figure(figsize=(16,8))
sns.countplot(data=df,x='embarked')
plt.xticks(rotation=90)
plt.xlabel("embarked")
plt.ylabel("Count")
```

```python
sns.heatmap(df.isnull(),cmap='viridis')
```

```python
sns.histplot( data = df, x='age', hue = 'sex', multiple = 'dodge')
```

```python
sns.histplot(df,x='age',color='orange')
plt.title("Passengers By Age")
plt.show()
```

```python
plt.figure(figsize=(16,8))
sns.countplot(data=df,x='survived')
plt.xticks(rotation=90)
plt.xlabel("survived")
plt.ylabel("Count")
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df,x='sex')
plt.xlabel('sex')
plt.ylabel('frequency')
```

```python
# Plotting the histogram of ticket prices
plt.figure(figsize=(10, 6))
sns.histplot(data=df,x='fare',stat='percent',bins=10, kde=True, color='skyblue')

plt.xlabel('Ticket Price')
plt.ylabel('Frequency')
plt.show()
```

```python
df[["sex", "survived"]].groupby(['sex'], as_index=False).mean().sort_values(by='survived', ascending=False)
```

```python
df[['pclass', 'survived']].groupby(['pclass'], as_index=False).mean().sort_values(by='survived', ascending=False)
```

```python
grid = sns.FacetGrid(df, col='survived', row='pclass', aspect=1.6)
grid.map(plt.hist, 'age', alpha=.5, bins=20)
grid.add_legend();
```
